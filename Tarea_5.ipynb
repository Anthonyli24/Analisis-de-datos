Nuestro objetivo es analizar cómo la pandemia ha afectado a la industria turística del país. Esta elección está motivada por la necesidad de comprender cómo los turistas extranjeros veían el atractivo cultural de Japón antes del brote y cómo se ve afectado hoy.

```python
# Dsitribución de porcentajes de visitantes a Japon por año
import pandas as pd
import matplotlib.pyplot as plt

def cantidadVisitoresXAnio():
    df = pd.read_csv('/content/Number of foreign visitors to Japan by month_ .csv')
    data = pd.DataFrame(df)

    anio2017 = data[data['Year'] == 2017]
    anio2018 = data[data['Year'] == 2018]
    anio2019 = data[data['Year'] == 2019]
    anio2020 = data[data['Year'] == 2020]
    anio2021 = data[data['Year'] == 2021]
    anio2022 = data[data['Year'] == 2022]
    anio2023 = data[data['Year'] == 2023]

    visitors_2017 = anio2017['Visitor'].sum()
    visitors_2018 = anio2018['Visitor'].sum()
    visitors_2019 = anio2019['Visitor'].sum()
    visitors_2020 = anio2020['Visitor'].sum()
    visitors_2021 = anio2021['Visitor'].sum()
    visitors_2022 = anio2022['Visitor'].sum()
    visitors_2023 = anio2023['Visitor'].sum()

    # Crear una lista con las sumas de visitantes por año
    visitantes_por_anio = [visitors_2017, visitors_2018, visitors_2019, visitors_2020, visitors_2021, visitors_2022,visitors_2023]

    # Etiquetas para los años
    etiquetas_anios = [2017, 2018, 2019, 2020, 2021, 2022,2023]

    # Crear un gráfico de pastel
    plt.figure(figsize=(10, 10))
    plt.pie(visitantes_por_anio, labels=etiquetas_anios, autopct='%1.1f%%', startangle=140)
    plt.title('Distribución de Visitantes por Año')

    # Mostrar el gráfico
    plt.show()

cantidadVisitoresXAnio()
```

Antes del brote,podemos observar que el atractivo de Japón entre los turistas extranjeros mostraba una importante tendencia al alza. Su enfoque en la rica cultura, su impresionante patrimonio histórico y su combinación única de tradición y modernidad lo convierten en un destino atractivo. Sin embargo, el número de turistas durante la epidemia (en 2020, 2021 y 2022) disminuyó significativamente, por lo que el desarrollo del turismo se detuvo repentinamente.

Aunque las restricciones de viaje comenzaron a disminuir en 2022, lo que permitió que el turismo se recuperara en cierta medida, Japón aún no ha recuperado completamente su volumen de visitantes anterior. Si bien el mundo para es 2023 volvio a un nivel significativo de normalidad, el país claramente enfrenta desafíos continuos para recuperar el turismo, posiblemente debido a que persiste una predisposición entre algunos viajeros que evitan destinos relacionados con el brote inicial, lo que ha contribuido a la recuperación lenta de Japón en términos de turismo internacional.

```python
#Grafico que nos muestra la cantidad de personas de Costa Rica que han visitado Japon en los últimos 6 años
import pandas as pd
import matplotlib.pyplot as plt

def cantidadVisitoresXAnio():
    df = pd.read_csv('/content/Number of foreign visitors to Japan by month_ .csv')
    data = pd.DataFrame(df)

    anios = [2017, 2018, 2019, 2020, 2021, 2022, 2023]
    visitantes_por_anio = []

    for anio in anios:
        visitantes_anio = data[(data['Year'] == anio) & (data['Country/Area'] == 'Costa Rica')]['Visitor'].sum()
        visitantes_por_anio.append(visitantes_anio)

    # Gráfico de barras
    plt.figure(figsize=(10, 6))
    plt.bar(anios, visitantes_por_anio, color='skyblue')
    plt.title('Distribución de Visitantes a Japón por Año (Costa Rica)')
    plt.xlabel('Año')
    plt.ylabel('Cantidad de Visitantes')
    plt.xticks(anios)
    plt.show()

cantidadVisitoresXAnio()
```

Al observar este gráfico, analizando las preferencias de los costarricenses, podemos observar una tendencia similar a la anterior, pero a menor escala. Antes de la pandemia, Japón se había convertido en un destino atractivo para los turistas costarricenses. Sin embargo, el interés en visitar el país disminuyó significativamente debido al brote inicial y los acontecimientos posteriores.
A pesar de la situación actual, que promete una recuperación gradual, los costarricenses están mostrando cierta cautela a la hora de reanudar plenamente los viajes a Japón, posiblemente debido a la conexión del país con el continente de inicio de la pandemia. Esta tendencia refleja la complejidad de la epidemia y su impacto duradero en los patrones de viajes internacionales.
